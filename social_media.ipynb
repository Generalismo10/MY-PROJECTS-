```python
import sys
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import plotly .express as px
pd.options.display.max_rows=100000
pd.options.display.max_columns=1000
df=pd.read_csv('social_media_data.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
a=df['likes'].mean()
b=df['shares'].mean()
c=df['comments'].mean()
d=df['views'].mean()
df['likes'].fillna(a,inplace=True)
df['shares'].fillna(b,inplace=True)
df['comments'].fillna(c,inplace=True)
df['views'].fillna(d, inplace=True)
```

```python
df.isnull().sum()
```

```python
# post_types against Likes
a=df.groupby('post_type')['likes'].sum().sort_values(ascending=False)
x=pd.array(df['likes'])
font1={'family':'serif', 'color':'yellow', 'size':15}
font2={'family':'serif', 'color':'yellow', 'size':15}
font2={'family':'serif', 'color':'yellow', 'size':15}
plt.title('Performance of likes against Comment',color='red',fontdict=font1)
plt.xlabel('post_type',color='blue', fontdict=font2)
plt.ylabel('likes',color='red', fontdict=font2)
plt.plot(a, color='blue', marker='3', ms=9, mec='hotpink',mfc='green',ls='--')
plt.show()
```

```python
# postypes against shares
b=df.groupby('post_type')['shares'].sum().sort_values(ascending=False)
x=pd.array(df['shares'])
font1={'family':'serif', 'color':'yellow', 'size':15}
font2={'family':'serif', 'color':'yellow', 'size':15}
font2={'family':'serif', 'color':'yellow', 'size':15}
plt.title('Performance of Post_Types against Shares',color='red',fontdict=font1)
plt.xlabel('post_type',color='blue', fontdict=font2)
plt.ylabel('shares',color='red', fontdict=font2)
plt.plot(b, color='blue', marker='3', ms=9, mec='hotpink',mfc='green',ls='--')
plt.show()
```

```python
# post types against Comment
c=df.groupby('post_type')['comments'].sum().sort_values(ascending=False)
x=pd.array(df['comments'])
font1={'family':'serif', 'color':'yellow', 'size':15}
font2={'family':'serif', 'color':'yellow', 'size':15}
font2={'family':'serif', 'color':'yellow', 'size':15}
plt.title('Performance of Post_types against Comments',color='red',fontdict=font1)
plt.xlabel('post_type',color='blue', fontdict=font2)
plt.ylabel('comments',color='red', fontdict=font2)
plt.plot(c, color='blue', marker='3', ms=9, mec='hotpink',mfc='green',ls='--')
plt.show()
```

```python
# post types against Comment
d=df.groupby('post_type')['views'].sum().sort_values(ascending=False)
x=pd.array(df['views'])
font1={'family':'serif', 'color':'yellow', 'size':15}
font2={'family':'serif', 'color':'yellow', 'size':15}
font2={'family':'serif', 'color':'yellow', 'size':15}
plt.title('Performance of Post_types against views ',color='red',fontdict=font1)
plt.xlabel('post_type',color='blue', fontdict=font2)
plt.ylabel('views',color='red', fontdict=font2)
plt.plot(d, color='blue', marker='3', ms=9, mec='hotpink',mfc='green',ls='--')
plt.show()
```

```python
# Pie Chart
a=df.groupby('post_type')['likes'].sum().sort_values(ascending=False)
x=pd.array(a)
thelabel=pd.array(df['post_type'])
myset=set(thelabel)
newlabel=myset
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'blue','size':20}
mycolor=pd.array(['yellow','pink','orange','blue','red','indigo'])
myexplode=pd.array([0,0,0.2])
plt.pie(x, labels=newlabel,colors=mycolor, autopct='%1.1f%%',labeldistance=0.73,pctdistance=0.6,explode=myexplode, startangle=45, shadow=True)
plt.title('Distribution of Total Likes by Post Type', fontdict=font1)
plt.legend(loc='upper left')
plt.show()
```

```python
# Pie Chart
a=df.groupby('post_type')['shares'].sum().sort_values(ascending=False)
x=pd.array(a)
thelabel=pd.array(df['post_type'])
myset=set(thelabel)
newlabel=myset
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'blue','size':20}
mycolor=pd.array(['yellow','pink','orange','blue','red','indigo'])
myexplode=pd.array([0,0,0.2])
plt.pie(x, labels=newlabel,colors=mycolor, autopct='%1.1f%%',labeldistance=0.73,pctdistance=0.6,explode=myexplode, startangle=90, shadow=True)
plt.title('Distribution of Total Shares by Post Type', fontdict=font1)
plt.legend(loc='upper left')
plt.show()
```

```python
# Pie Chart
a=df.groupby('post_type')['views'].sum().sort_values(ascending=False)
x=pd.array(a)
thelabel=pd.array(df['post_type'])
myset=set(thelabel)
newlabel=myset
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'blue','size':20}
mycolor=pd.array(['yellow','pink','orange','blue','red','indigo'])
myexplode=pd.array([0,0,0.2])
plt.pie(x, labels=newlabel,colors=mycolor, autopct='%1.1f%%',labeldistance=0.73,pctdistance=0.6,explode=myexplode, startangle=90, shadow=True)
plt.title('Distribution of Total views by Post Type', fontdict=font1)
plt.legend(loc='upper left')
plt.show()
```
